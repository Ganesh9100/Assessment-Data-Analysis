**4.Find all pass students , 1 subject fail students, 5 subject fail student**

```
"""
 This function will replace the marks
 got by students in to 2 Categorie's 
 Either Pass or Fail

 PASSING CRITERIA :
 
 PASS: Students whose Mark is above 45 
 FAIL: Student's who got 45 and  below .

 Args:
  df_ : DataFrame variable 

 Return:
  None
  """

def criteria(df_):
  

  for i in range(10):         # for iterating over all rows
    for j in range(1,6):      # for iterating over all columns
      if df_.iloc[(i,j)]>45:  # Condition for Passing
        df_.iloc[(i,j)]='P'
        
      else:
        df_.iloc[(i,j)]='F'   
  
criteria(df_)     
```
