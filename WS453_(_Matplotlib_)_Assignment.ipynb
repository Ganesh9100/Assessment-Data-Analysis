7. **Write a Python program to create multiple plots. (Use subplot).**

```
# Creating Data Points

y3=[50, 16, 28, 17, 48]
x3=[ 31, 32, 33, 34, 35]


# Making Sub-Plot of (2,2)
fig, axs = plt.subplots(2,2)

# Axis 0,0
axs[0, 0].plot(x_day, y_wind)
axs[0, 0].set_title('Day-Wind')

# Axis 0,1
axs[0, 1].plot(x_day, y_temp, 'tab:orange')
axs[0, 1].set_title('Day-Temperature')

#Axis 1,0
axs[1, 0].plot(y_temp,[1,2,3,4,5,6,7,8,9], 'tab:green')
axs[1, 0].set_title('Wind-Temperature')

#Axis 1,1
axs[1, 1].plot(x3,y3, 'tab:red')
axs[1, 1].set_title('Simple')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
